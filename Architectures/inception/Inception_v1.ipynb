# inception v1
- Inception network was once considered a state-of-the-art deep learning architecture (or model) for solving image recognition and detection problems. 
> First, we need to be clear that the most direct way to improve the performance of deep neural networks is to increase the depth and width, but this will bring two problems:

- 1. A larger size usually means more parameters, which makes the larger network easier to overfit, especially when the labeled samples of the training set are limited.
- 2. Will consume a lot of computing resources.

**The design philosophy of GoogLeNet is:** 

- 1.The protruding(sticking out) part in the image may have a great dimensional change.

- 2.This huge change in information location makes it difficult to select the correct kernel size for convolution operations.

- 3.For more globally distributed information, larger cores are preferred, and for more locally distributed information, smaller cores are preferred.

- 4.Very deep networks are easy to overfit.It is also difficult to transfer gradient updates through the entire network.

- 5.Simple stacking of large convolution operations leads to high computational complexity

- GoogLeNet v1 contains 22 layers in total, and the network parameter amount is only 1/12 of Alexnet. With the deepending of the network layer, there will still be the problem of gradient disappearance, so the author adds two auxiliary softmax in the middle layer, in order to increase the size of the gradient of back propagation, and also plays a role in regularization. When calculating the network loss, the auxiliary softmax loss in the middle will be multiplied by a weight (0.3) and added to the loss value of the last layer. During prediction, the output of the middle softmax layer is ignored.

# inception_v2 
- In the Inception-v2, they introduced Factorization(factorize convolutions into smaller convolutions) and some minor change into Inception-v1.

-**The improvements of Inception v2 are mainly in the following points:**

- 1. JoinBatch Normalization(Batch normalization) layer, the standard structure is: Convolution-BN-relu

- 2. Drawing on the use of VGG, two 3 * 3 convolutions are used in series to replace the 5 * 5 convolution module in the Inception module. Because two 3 * 3 convolutions have the same receptive field as a 5 * 5 convolution, but the amount of parameters is less than 5 * 5 convolutions. And because of the addition of a layer of convolution operation, there is one more Relu, that is, a layer of nonlinear mapping is added, which makes the feature information more discriminative.

- 3. Use asymmetric convolution to further decompose the 3 * 3 convolution into 3 * 1 and 1 * 3 convolutions. As shown below:

- 1) The first convolution is performed by n × 1 and then by 1 × n convolution, which is equivalent to the result of directly performing n × n convolution.

- 2) Asymmetric convolution can reduce the amount of calculation. It turned out to be n × n multiplications. After the decomposition, it became 2 × n multiplications. The larger n is, the more the amount of calculation decreases.

- 3) Although it can reduce the amount of calculation, this method is not applicable everywhere. Such an asymmetric convolution should not be used in the layer close to the input, which will affect the accuracy. It should be used in the higher layer.Asymmetric convolution works better when the image size is between 12 × 12 and 20 × 20.

# Inception v3
- In the Inception-v3 also, they introduced Factorization(factorize convolutions into smaller convolutions) and some minor change into Inception-v1.
- As for Inception-v3, it is a variant of Inception-v2 which adds BN-auxiliary.
- Inception v3 adopts Inception v2's network structure as a whole, and has made improvements in optimization algorithms and regularization, as follows:  

> 1. Optimization algorithm uses RMSProp to replace SGD

> 2. BN operation(batch normalization operation) is also added to the auxiliary classifier(auxiliary classifier is used to tackle the vanishing gradient problem after that loss of auxiliary classifier is added to the loss of original classifier to increase the performance of the model)

>3. Use the Label Smoothing Regularization (LSR) method to avoid overfitting, prevent the network from over-predicting a certain category, and add more attention to small probability categories. (For the LSR optimization method, you can refer to this blog https://blog.csdn.net/lqfarmer/article/details/74276680)

>4. Replace the first 7x7 convolution kernel with two 3x3 convolution kernels

>BN auxiliary refers to the version in which the fully connected layer of the auxiliary classifier is also-normalized, not just convolutions. We are refering to the model [Inception-v2 + BN auxiliary] as Inception-v3.

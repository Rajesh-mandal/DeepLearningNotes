# DeepLabv2
- DeepLab: Semantic Image Segmentation with Deep Convolutional Nets, Atrous Convolution, and Fully Connected CRFs 
- DeepLabv2 is an architecture for semantic segmentation that build on DeepLab with an atrous spatial pyramid pooling scheme. Here we have parallel dilated convolutions with different rates applied in the input feature map, which are then fused together   
## introduction
DCNN has three challenges in semantic segmentation:    
> (1) Feature resolution decreases (2) There are multiple scales of objects (3) Due to the spatial invariance of DCNN, the spatial accuracy decreases
- (1) Because the repeated pooling and downsampling in DCNN reduce the spatial resolution, one method is to use deconvolutional layer (deconvolutional layer), but it requires additional space and calculation. DeepLabv2 replaces downsampling with hole convolution in the last few maximum pooling layers, and calculates the feature map with a higher sampling density.
- (2) The object has multiple scales. A standard way to solve this problem is to scale the image to different sizes, and summarize the features to get the result. This method can improve performance, but it increases computational cost. Inspired by SPPNet, DeepLabv2 proposes a similar structure, which samples a given input with different sampling rates of the hole convolution, and captures the image context in multiple proportions, which is called ASPP (astrous spatial pyramid pooling).
- (3) The classifier requires spatial invariance, which limits the spatial accuracy of DCNN. One way to solve this problem is to use the skip structure to fuse the features of different layers to calculate the final segmentation result. A more efficient method of DeepLabv2 is to use conditional random fields to enhance the model's ability to capture details.


**DeepLabv2 contribution** 

Using multi-scale processing and ASPP to achieve better performance.   

On the basis of DeepLab, VGG-16 was replaced with ResNet, and SOTA was achieved on PASCAL VOC 2012 and other data.

## Summary
Commonly used structures in semantic segmentation models: (a) spatial pooling pyramid, (b) encoding-decoding structure. The ASPP in deeplab-v3 uses hole convolution with different sampling rates and global pooling operations to encode multi-scale context information; the encoding-decoding structure gradually restores spatial detail information through the decoding process to obtain a clear target boundary. In deeplabv3+, the advantages of the two are combined to achieve better performance.   

- Use the improved Xception as the backbone
- Use depth separable convolution in ASPP and decoder modules to reduce parameters and increase speed

## Decoder part:
- 11 Conv: Select the feature map (OS=4) with a resolution of 1/4 of the original image from DCNN as Low-Level Features. Since the number of channels of low-level features is usually relatively large, the number of high-level feature channels output by the encoder is only 256. Avoid low-level features. The network considers them more important due to the large number of channels, so the number of channels is reduced by 11 convolution. And it is proved through experiments that mIOU is the highest when the number of channels is 48  


- 33 Conv: After fusing the low-level features and the high-level features output by the encoder, the concat feature map is then integrated through 33 convolution, and its configuration is also determined through experiments
